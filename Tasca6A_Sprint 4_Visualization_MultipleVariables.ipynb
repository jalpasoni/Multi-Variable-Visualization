### Correlation matrix

```python
fig = plt.figure(figsize=(18, 8))
corr = dummies.corr()
kot = corr[corr>=0.2]
filteredDf = corr[((corr >= .2) | (corr <= -.2)) & (corr !=1.000)]
sns.heatmap(filteredDf, cmap="coolwarm", annot=True)
plt.grid('on')
```

    The following function compares the dependency of two movie genres with eachother, that is, the likelyhood of a movie being classified into one also being classified into the other. 

```python
dummies.columns
```

```python
def gen_compare(g1, g2):
    genre_crosstab = pd.crosstab(dummies[g1], dummies[g2], margins = False) 
    #print(genre_crosstab)
    a1 = genre_crosstab[0][0]/(genre_crosstab[0][0]+genre_crosstab[1][0])
    a2 = genre_crosstab[0][1]/(genre_crosstab[0][1]+genre_crosstab[1][1])
    dependency = "{:.0%}".format(abs(a1-a2))
    print("Dependency between %s and %s is: %s" % (g1, g2, dependency))

g1 = 'Animation'
g2 = 'Children\'s'

gen_compare(g1, g2)
```

```python
g1 = 'Comedy'
g2 = 'Fantasy'

gen_compare(g1, g2)
```

```python
import matplotlib.colors as mcolors

colors2 = plt.cm.binary(np.linspace(0.2, 0.8, 128))
colors1 = plt.cm.gist_heat_r(np.linspace(0.01, 0.5, 128))
# combine them and build a new colormap
colors = np.vstack((colors1, colors2))
mymap = mcolors.LinearSegmentedColormap.from_list('my_colormap', colors)

sns.set(rc={'figure.figsize':(15,10)})
sns.heatmap(movies_genre_pivot, linewidths=.5, cmap=mymap);
```
